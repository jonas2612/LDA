```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel("Hwk data sets/Data set Toenail/toenail_with_missings.xls")
data
```

```python
t1 = data[['id', 'age']].groupby(['id']).mean()
t2 = data[['id', 'age']].groupby(['id']).max()

all(t1['age'] == t2['age'])
```

```python
for id in np.unique(data.id):
    data[data.id==id].age = min(data[data.id==id].age)
    
data.id = pd.Categorical(data.id)
data.health_club = pd.Categorical(data.health_club)
data.gender = pd.Categorical(data.gender)
data.treatment = pd.Categorical(data.treatment)
data.dtypes
```

```python
grid = sns.FacetGrid(data, col="gender", row="treatment")
grid.map(sns.lineplot, "month", "Unafflicted_nail_length_with_NA", hue=data.id, style=data.health_club)
```

```python
grid = sns.FacetGrid(data, col="gender", row="treatment", hue="id")
grid.map(sns.lineplot, "month", "Unafflicted_nail_length_with_NA")#, style=data.health_club)
```

```python
grid = sns.FacetGrid(data, col="gender", row="health_club")
grid.map(sns.lineplot, "month", "Unafflicted_nail_length_with_NA", hue=data.id, style=data.health_club)
```

```python
grid = sns.FacetGrid(data, col="health_club", row="treatment")
grid.map(sns.lineplot, "month", "Unafflicted_nail_length_with_NA", hue=data.id, style=data.gender)
grid.add_legend()
```

```python
grid = sns.FacetGrid(data, col="gender", row="age")
grid.map(sns.lineplot, "month", "Unafflicted_nail_length_with_NA", hue=data.id, style=data.health_club)
```
